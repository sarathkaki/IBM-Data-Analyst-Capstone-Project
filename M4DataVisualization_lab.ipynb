## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# SQL query to retrieve the 'Age' and 'WorkWeekHrs' values from the master table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Read the data into a DataFrame using pandas
df_age_workhrs = pd.read_sql_query(QUERY, conn)

# Create a scatter plot using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df_age_workhrs)
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# SQL query to retrieve the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' values from the master table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Read the data into a DataFrame using pandas
df_work_age_code = pd.read_sql_query(QUERY, conn)

# Create a bubble plot using seaborn
plt.figure(figsize=(12, 8))
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(30, 300), data=df_work_age_code)
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# SQL query to retrieve the median 'WorkWeekHrs' and 'CodeRevHrs' for the age group 30 to 35
QUERY = """
SELECT Age,
       (SELECT AVG(WorkWeekHrs)
        FROM master
        WHERE Age BETWEEN 30 AND 35) as MedianWorkWeekHrs,
       (SELECT AVG(CodeRevHrs)
        FROM master
        WHERE Age BETWEEN 30 AND 35) as MedianCodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""

# Read the data into a DataFrame using pandas
df_age_group = pd.read_sql_query(QUERY, conn)

# Create a stacked chart using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='MedianWorkWeekHrs', data=df_age_group, color='blue', label='WorkWeekHrs')
sns.barplot(x='Age', y='MedianCodeRevHrs', data=df_age_group, color='orange', label='CodeRevHrs')

plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()
plt.show()
```

```python
# SQL query to retrieve the median 'ConvertedComp' for ages from 25 to 30
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
"""

# Read the data into a DataFrame using pandas
df_age_group = pd.read_sql_query(QUERY, conn)

# Calculate median for each age group
age_range = range(25, 31)  # Include 30 in the range
median_comp = [df_age_group[df_age_group['Age'] == age]['ConvertedComp'].median() for age in age_range]

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(age_range, median_comp, marker='o', linestyle='-', color='b')

plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (US$)')
plt.title('Line Chart of Median ConvertedComp for Ages 25 to 30')
plt.grid(True, linestyle='--')

plt.show()
```

```python
correlation_coefficient = df_age_group['Age'].corr(df_age_group['ConvertedComp'])
print("Pearson Correlation Coefficient:", correlation_coefficient)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# SQL query to retrieve the median 'ConvertedComp' for ages from 45 to 60
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE age <= 60 AND age >= 45
ORDER BY age
"""

df = pd.read_sql_query(QUERY, conn)

# Calculate median for each age group
l1 = range(45, 61)  # Use range(45, 61) to include 60 in the range
l2 = [df[df['Age'] == i]['ConvertedComp'].median() for i in l1]

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(l1, l2, marker='o', linestyle='-', color='b')

plt.xlabel('Age')
plt.ylabel('Converted Compensation (US$)')
plt.title('Line Chart of Median ConvertedComp for Ages 45 to 60')
plt.grid(True, linestyle='--')

plt.show()
```
